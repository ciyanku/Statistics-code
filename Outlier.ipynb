```python
import numpy as np
```

```python
arr=np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
arr
```

```python
Q1=np.percentile(arr,25)
Q3=np.percentile(arr,75)
IQR=Q3-Q1
print(IQR)
```

```python
upper=Q3+1.5*IQR
Lower=Q1-1.5*IQR
```

```python
upper
```

```python
Lower
```

```python
l=[]
for i in arr:
    if i>=Lower and i<=upper:
        l.append(i)
```

```python
arr2=np.array(l)
arr2
```

```python
import seaborn as sns
sns.boxplot(arr2)
```
